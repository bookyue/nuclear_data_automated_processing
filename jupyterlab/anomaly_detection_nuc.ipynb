# Getting The Data

```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from nuc_data_tool.db.fetch_data import fetch_files_by_name, fetch_data_by_filename_and_physical_quantities
files = fetch_files_by_name()
```

```python
import pandas as pd

nuc_data = pd.DataFrame(index=['nuc_ix', 'name'])

nuc_list = ['U233', 'Ce142', 'Tc99', 'Nd144', 'I129', 'U238', 'Gd155', 'Am241', 'Pm149', 'Am242', 'Pr144', 'La139', 'Ba138', 'Np239', 'U236', 'I135', 'Cm242', 'Sm150', 'Sm149', 'Sn126', 'Y91', 'Np237', 'Eu153', 'Cm244', 'Pu239', 'Se79', 'Kr83', 'Xe136', 'Ag109', 'U234', 'Xe135', 'Ru101', 'Mo95', 'Pa233', 'U239', 'Sr90', 'Cs133', 'Sm147', 'Cs135', 'U235', 'Zr94']
for file in files:
    dict_nuc_data = fetch_data_by_filename_and_physical_quantities(file, 'isotope', True)

    for pq in dict_nuc_data:
        if dict_nuc_data[pq].empty:
            continue

        need_ix = dict_nuc_data[pq]['name'].isin(nuc_list)
        nuc_data_needed = dict_nuc_data[pq][need_ix].copy()
        nuc_data_needed.rename(columns={'first_step': f'{file.name}_first_step',
                                        'last_step': f'{file.name}_last_step'},
                               inplace=True)
        columns = {col: f'{file.name}_{col}'
                   for col in nuc_data_needed.columns.tolist()
                   if 'middle_step' in col}
        nuc_data_needed.rename(columns=columns, inplace=True)

        nuc_data = pd.merge(nuc_data.T, nuc_data_needed, how='outer', on=['nuc_ix', 'name']).T
```

```python
nuc_data.columns = nuc_data.iloc[1]

nuc_data.drop(['nuc_ix', 'name'], inplace=True)

# nuc_data.reset_index(drop=True, inplace=True)
```

# Setting up Environment
